```python
import numpy as np
import pandas as pd
```

```python
import keras
```

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

```python
#Let's check how many data points we have.
concrete_data.shape
```

```python
#Let's check the dataset for any missing values with describe.
concrete_data.describe()
```

```python
#Let's check the dataset for any missing values with describe.
concrete_data.isnull().sum()
```

```python
# it stores the column names or attribute name in list
concrete_data_columns = concrete_data.columns
concrete_data_columns
```

```python
# Divide the dataset into predictors and target sub datasets. It just a slicing on the dataset
# Predictors Dataset is  a dataframe
predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
predictors.head()
```

```python
# the target dataset is a dataframe
target = concrete_data['Strength'] # Strength column
target.head()
```

```python
# The last step is to normalize the data by substracting the mean and dividing by the standard deviation.
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
#Let's save the number of predictors to n_cols since we will need this number when building our network.
n_cols = predictors_norm.shape[1] # number of predictors
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
#Build a Neural Network
```

```python
#Let's define a function that defines our regression model for us so that we can conveniently call it to create our model.
```

```python
def regression_model():
    #creating model
    model=Sequential()
    #adding layers with input dimension, no. of neurons and activation function
    # first hidden layer
    model.add(Dense(50, activation='relu', input_shape=(n_cols,)))
    #second hidden layer
    model.add(Dense(50, activation = 'relu'))
    # output layer
    model.add(Dense(1))
    
    #compile the model
    model.compile(optimizer ='adam', loss='mean_squared_error')
    return model
```

```python
#Train and Test the Network
```

```python
#Let's call the function now to create our model.
```

```python
# build the model
model = regression_model()
```

```python
#fit the model
model.fit(predictors_norm, target, validation_split=0.3, epochs=100, verbose=2)
```
